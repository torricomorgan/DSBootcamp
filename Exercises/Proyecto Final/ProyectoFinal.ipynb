```python
reason_columns = pd.get_dummies(df["Reason for Absence"], drop_first=True)
```

```python
df = df.drop(["Reason for Absence"], axis=1)
```

```python
df
```

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'TransportationExpense', 'DistanceToWork', 'Age',
       'DailyWorkLoadAverage', 'BodyMassIndex', 'Education',
       'Children', 'Pets', 'AbsenteeismTimeInHours', "Reason_1", "Reason_2", "Reason_3", "Reason_4"]
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
column_names_reorder = ["Reason_1", "Reason_2", "Reason_3", "Reason_4", 'Date', 'TransportationExpense', 'DistanceToWork', 'Age',
       'DailyWorkLoadAverage', 'BodyMassIndex', 'Education',
       'Children', 'Pets', 'AbsenteeismTimeInHours']
```

```python
df = df[column_names_reorder]
df.head()
```

# Checkpoint

```python
df_checkpoint = df.copy()
```

# Date

```python
df_checkpoint["Date"]
```

```python
df_checkpoint["Date"] = pd.to_datetime(df_checkpoint["Date"], format = "%d/%m/%Y")
```

```python
df_checkpoint["Date"]
```

```python
df_checkpoint.info()
```

```python
df_checkpoint["Date"][0].month
```

```python
list_months = []
```

```python
for i in range(df_checkpoint.shape[0]):
	list_months.append(df_checkpoint["Date"][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_checkpoint["Month Value"] = list_months
```

```python
df_checkpoint.head(10)
```

```python
def date_to_weekday(date_value):
	return date_value.weekday()
```

```python
df_checkpoint["DayOfTheWeek"] = df_checkpoint["Date"].apply(date_to_weekday)
```

```python
df_checkpoint.head(10)
```

```python
df_checkpoint.columns.values
```

```python
df_checkpoint = df_checkpoint.drop(["Date"], axis = 1)
df_checkpoint.head()
```

```python
columns_reorder2 = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'DayOfTheWeek',
       'TransportationExpense', 'DistanceToWork', 'Age',
       'DailyWorkLoadAverage', 'BodyMassIndex', 'Education', 'Children',
       'Pets', 'AbsenteeismTimeInHours']
```

```python
df_checkpoint = df_checkpoint[columns_reorder2]
df_checkpoint.head()
```
